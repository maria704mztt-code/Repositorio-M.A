**ARREGLOS** son los elementos primordiales de Numpy, los cuals son colecciones ordenadas e indexables de datos. (np.array)

```python
import numpy as np
arreglo = np.array([1,2,3,4], dtype = float)
print(arreglo)
```

**DIMENSIONES Y FORMAS DE ARREGLO **
Los arreglos pueden tener arreglos en varios niveles o dimensiones. para observar sus dimensiones y formas se crea el siguiente arreglo: arreglo = np.array([1,2,3],[4,5,6]). ndim indica en numero de dimensiones del arreglo.

```python
print("el numero de dimensiones del arreglo es : ", arreglo.ndim)
```

Todos los arreglos en una dimension especifica deben de contener el mismo numero de elementos

```python
print("la forma del arreglo es:", arreglo.shape)
print("el numero total de elementos del arreglo es:", arreglo.size)
print("los tipos de datos del arreglo son:", arreglo.dtype)
```

shape: describe el numero de elemntos que contiene cada dimension del arreglo.
size: indica el numero total de elementos que conforman un arreglo.
dtype: indica el tipo de dato de un arreglo.

```python
arreglo2d = np.array([[1,2,3],[4,5,6]],dtype=float)
print(arreglo2d)
```

```python
print("el numero de dimensiones del arreglo es : ", arreglo2d.ndim)
print("la forma del arreglo es:", arreglo2d.shape)
print("el numero total de elementos del arreglo es:", arreglo2d.size)
print("los tipos de datos del arreglo son:", arreglo2d.dtype)
```

```python
arreglo3d = np.array([
    [[1,2,3],[5,6,7]],
    [[7,8,9],[10,11,12]]
])
arreglo3d
```

```python
print("el numero de dimensiones del arreglo es : ", arreglo3d.ndim)
print("la forma del arreglo es:", arreglo3d.shape)
print("el numero total de elementos del arreglo es:", arreglo3d.size)
print("los tipos de datos del arreglo son:", arreglo3d.dtype)
```
